# Задание 1

```python
word = input('Введите слово:\n ')

if len(word)%2 == 1:
    print(f"Средняя буква слова: {word[len(word)%2]}")
else:
    print(f"Среднии буквы слова: {word[(len(word)//2)-1:(len(word)//2)+1]}")
```

# Задание 2

```python
a = int(input('Введите число:\n '))
summ_ = 0

while a != 0:
    summ_ += a
    a = int(input('Введите число:\n '))
else:
    print(f"Result: \n{summ_}")
```

# Задание 3

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys)==len(girls):
    l = list(zip(sorted(boys), sorted(girls)))

    print("Результат: \nСомнительно идеальные пары: ")
    for pair in l:
        print(f"{pair[0]} и {pair[1]}")

else:
    print("Результат: \nВнимание, кто-то может остаться счастливчиком!")
```

# Задание 4

```python
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
print('Средняя температура в странах:')
for country in countries_temperature:
    print(f"{country[0]} - {round(((sum(country[-1])/len(country[-1]))-32)*5/9, 1)} C")
```

# Задание 6

```python
nums = input('Введите числа через пробел: \n')
repeat = []
alls = []

num_dict = nums.split()

for num in num_dict:
    if num not in alls:
        alls.append(num)
    elif num in repeat:
        pass
    else:
        repeat.append(num)
print(sorted(repeat))
```
