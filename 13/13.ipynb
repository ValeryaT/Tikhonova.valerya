```python
import matplotlib.pyplot as plt
import pandas as pd

from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
df = pd.read_csv('2019.csv')

data = df[['Country or region', 'Score', 'GDP per capita',  'Social support',
            'Healthy life expectancy', 'Freedom to make life choices', 'Generosity',
            'Perceptions of corruption']]
data.columns = ['Country', 'Score', 'GDP',  'SocSup', 'HLE',
                             'Freedom', 'Generosity', 'Corruption']
data.head()
```

```python
data['Corruption'].plot(kind='box', title='Уровень коррупция за 2019')
```

График уровня коррупции показывает, что большинство стран успешно борется с коррупцией, но есть и небольшое кол-во стран, в которых этот коэффициент находится на высоком уровне



```python
data.plot(kind='scatter', y='Corruption', x='Score',
        title='График зависимости коррупции от уровня счастья в стране')
plt.show()
```

Из графика зависимости коррупции от уровня счастья в стране видно, что в странах со средним уровнем счастья коэф. коррупции с низкими показателями. Чем выше уровеь счастья, тем выше показатели коррупции. хм....

```python
d = data[['Score',  'HLE', 'Corruption']]
d.plot(kind='bar',
                    xlabel='Страны',
                    ylabel='Коэффициенты')
```

Интересная зависимость уровня ЗОЖ от счастья. Уровень ЗОЖ гораздо стабильнее и наиболее высок, когда уровень счастья имеет наибольшие показатели. Это значит, что счастливые люди больше ценят свое здоровье)

```python
r= data[ (data['Country']=='Russia')]
other = data[ (data['Country']=='Armenia') | (data['Country']=='Azerbaijan')
        | (data['Country']=='Artsakh') | (data['Country']=='Belarus') | (data['Country']=='Estonia')
        | (data['Country']=='Georgia') | (data['Country']=='Abkhazia') | (data['Country']=='South Ossetia')
        | (data['Country']=='Kazakhstan') | (data['Country']=='Kyrgyzstan') | (data['Country']=='Latvia')
        | (data['Country']=='Lithuania') | (data['Country']=='Moldova') | (data['Country']=='Transnistria')
        | (data['Country']=='Tajikistan') | (data['Country']=='Turkmenistan') | (data['Country']=='Ukraine')
        | (data['Country']=='Uzbekistan')]
ussr = [other, r]

for one in ussr:
    plt.bar(one['Country'],one['Score'])
    plt.title('Уровень счастья стран бывшего Советского Союза 2019')
    plt.xlabel('Country')
    plt.ylabel('Score')
    plt.legend(['USSR', 'Russia'])


```

Из графика видно, что неожиданным лидером по индексу счастья является Узбекистан, а антилидер - Украина (тут без неожиданностей). Россия же находится примерно в середине списка бывших стран СССР. Наверное, нужно сгонять в Узбекистан.
